```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
#Load Iris dataset
iris = datasets.load_iris()
iris
```

```python
#let get our target name
iris.target_names
print("Target names:", iris.target_names)
```

```python
 # let get our features names
 iris.feature_names
 print("Feature names:", iris.feature_names)
```

```python
#The dtype of the features
print("Feature data type:", iris.data.dtype)
```

```python
import pandas as pd
# Create DataFrame with features
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add target column
df['target'] = iris.target
```

```python
# size of the data
df.shape
```

```python
df.head()
```

```python
df.tail()
```

```python
#Assign each name a number starting from 0
for i, name in enumerate(iris.target_names):
    print(f"{i} = {name}")
```

```python
#Split data into training (75%) and testing (25%)
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target  , test_size=0.25, random_state=42
)
```

```python
#Display sizes of training samples and labels
print("Training data shape:", X_train.shape)
print("Training labels shape:", y_train.shape)
```

```python
# using the KNN classifier model
knn = KNeighborsClassifier(n_neighbors=1)
# train the model
knn.fit(X_train, y_train)
```

```python
# let make some prediction
new_data = np.array([[5.1, 3.5, 1.4, 0.2]])
prediction = knn.predict(new_data)
print("Predicted class:", iris.target_names[prediction][0])
```

```python
new_data1 = np.array([[4.8, 2.3, 1.4, 1.0]])
prediction = knn.predict(new_data1)
print("Predicted class:", iris.target_names[prediction][0])
```

```python
new_data2 = np.array([[6.8, 3.3, 5.4, 2.0]])
prediction = knn.predict(new_data2)
print("Predicted class:", iris.target_names[prediction][0])
```

```python
accuracy = knn.score(X_test, y_test)
print("Model accuracy:", accuracy)
```

##Interpretation of Results

The model performs very well (close to 100% accuracy).

Using 1 neighbor means the prediction is based on the single closest sample.

Iris dataset is well-separated, so KNN works extremely well.

However, using k=1 can sometimes cause overfitting (too sensitive to noise).
