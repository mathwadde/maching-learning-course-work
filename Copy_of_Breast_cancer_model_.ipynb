```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

data = load_breast_cancer()

X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target)

print(X.head())
print(y.value_counts())
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.hist(y)
plt.show()
```

```python
X.shape
```

```python
df = pd.concat([X, y], axis=1)
df.head(20)
```

```python
df.columns
```

```python


#  inplace rename
df= df.rename(columns={0: "binary"})
df.columns
```

```python
df = df[['mean area','mean concave points','compactness error','mean symmetry','worst fractal dimension', 'binary'  ]]
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# let have a df of dataset without 'binary' as target and check correlation with heat map
corr_matrix = df.drop(columns='binary').corr()

plt.figure(figsize=(12,10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.show()
```

The heat map shows that the mean concave point and mean area are correlated, so we have to drop one. we drop drop mean concave point because  the mean area dataset give a better explanantory variable  

```python
df = df.drop(columns=['mean concave points'])
df.columns
```

```python
df =  df.drop_duplicates()
df.head()
```

```python
# let check vif test
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_data = pd.DataFrame()
vif_data["feature"] = df.drop(columns='binary').columns
vif_data
```

from the VIF test all our features are lesser than 5, so they make sense. Thus we can run a model

```python
# let  split the data set
from sklearn.model_selection import train_test_split
feature = df.drop(columns='binary')
target = df['binary']
X_train, X_test, y_train, y_test = train_test_split(feature, target, test_size=0.2, random_state=42)

# let instantiate the model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
#let fit the model
model.fit(X_train, y_train)
# let prdict result
y_pred = model.predict(X_test)
```

```python
# let compute the model accuracy
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

```

```python
# we can compare cross_validation score with R2
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(model, feature, target, cv=5)
print("Cross-Validation Scores:", cv_scores)
print("Mean CV Score:", cv_scores.mean())
```

```python
# let show our cross validation on a line curve
import matplotlib.pyplot as plt
plt.plot(cv_scores)
```

```python
# let create a reinforced model_1
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
model_1 = LogisticRegression()
model_1.fit(X_train_scaled, y_train)
y_pred_1 = model_1.predict(X_test_scaled)
accuracy_1 = accuracy_score(y_test, y_pred_1)
print("Accuracy with Scaling:", accuracy_1)


```

```python
#LET MAKE SOME FURTHER PREDICTIONS
Y_PRED = model_1.predict(X_test_scaled)
print(Y_PRED)
```

```python
# let see the graph of the two models by comparing their accuracy
import matplotlib.pyplot as plt
models = ['Original Model', 'Model with Scaling']
accuracies = [accuracy, accuracy_1]
plt.bar(models, accuracies)
plt.xlabel('Model')
plt.ylabel('Accuracy')
```

```python
# LET EXTRACT FEAUTURE NAMES AND IMPORTANCE
feature_names = feature.columns
importances = model_1.coef_[0]
# LET SHOW THEM IN A SERIES
feature_importance = pd.Series(importances, index=feature_names)
print(feature_importance)
```

```python
#let plot a horizontal bar chart for the feature importance
import matplotlib.pyplot as plt
col = ['green', 'blue','green', 'green']
plt.barh(feature_names, importances, color = col)
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importance')
```

```python
# LET DO SOME ODD-RATIO
import numpy as np
odds_ratio = np.exp(feature_importance)
print(odds_ratio)
```

```python
# LeT PlOT bar chart
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 5))
plt.xticks(rotation=60, fontsize=8)
plt.bar(feature_names, odds_ratio)
plt.xlabel('Feature')
plt.ylabel('Odds Ratio')
;
```
